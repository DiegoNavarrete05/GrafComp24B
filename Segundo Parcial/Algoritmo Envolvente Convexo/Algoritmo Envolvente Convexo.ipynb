### Este código genera un conjunto de puntos aleatorios en segunda dimension y calcula el "convex hull" (envolvente convexa) de estos puntos de manera que el polígono más pequeño que puede encerrar todos los puntos.

* 1. Se generan los puntos aleatorios en un rango determinado.
  2. Se usa la funcion convex_hull() para calcula la envolvente convexa dividiendo el conjunto en dos partes, superior e inferior.
  3. Usando la función turn_right() que utiliza el determinante para mantener la orientación correcta del polígono.
