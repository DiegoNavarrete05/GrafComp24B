```python
from sympy import Matrix

# Define la matriz A
A = Matrix([[1, 1, 0],
            [1, 1, 0]])

# Calcula el núcleo (nullspace)
nucleo = A.nullspace()

# Imprime el núcleo
print("El núcleo de la matriz es:")
for vector in nucleo:
    print(vector)
```

```python
from sympy import Matrix

# Define la matriz A
A = Matrix([[1, 1, 2],
            [2, 2, 4],
            [2, 3, 5]])

# Calcula el núcleo (nullspace)
nucleo = A.nullspace()

# Imprime el núcleo
print("El núcleo de la matriz es:")
for vector in nucleo:
    print(vector)
```
